This is a programme that will generate graphs based on the CSV file generated by the rtl_power command prompt utility that comes with RTL-SDR devices. In this instance, we are using the NooElec RTL-SDR V5 with the SAWBird+ H1. This code will generate an overall reading based on the output and then generate a graph specific to the hydrogen line frequency. The syntax of the command I've used is:

"rtl_power -f 1419.5M:1420.5M -g 50 -O -h 4h radiotelescope.csv"

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot
reading=pd.read_csv("radiotelescope.csv",header=None) # import the csv file. Place the CSV file in the same folder as the
                                                        # code file. The header=None denotes that the CSV file does not
                                                        # contain column headings.
```

```python
reading.head()
```

The CSV file does not have column headings, so we have to assign them. The first 6 columns contain specific variables. They are the date of the reading, the time, the start frequency, the end frequency, the frequency step and number of samples. The rest of the columns are the readings themselves split into individual frequencies

Each column represents the Start Frequency + the Frequency Step * the column number, starting with 1. The code below assigns each column a name.

```python
start_freq=reading.at[1,2]
freq_div=reading.at[1,4]
reading_columns=["Date_Reading","Time_Reading","Start_Freq","Stop_Freq","Freq_Div","Sample_Size",str(start_freq)]
no_of_columns=(len(reading.axes[1]))-7
new_columns=[str(start_freq)]
i=1
while i<=no_of_columns:
    next_column_name=(start_freq+(freq_div*i))
    reading_columns.append(str(next_column_name))
    i += 1
reading.columns=reading_columns
```

```python
reading.head()
```

In order to generate the graphs, we need to collapse the readings columns into rows labelling them "Frequency" and "Value". We use the df.melt utility for this. We will also drop the Date, low and high frequency and sample size variables as they are superflous for graphing purposes.

```python
readings_converted=reading.melt(["Time_Reading", "Date_Reading","Start_Freq","Stop_Freq","Freq_Div","Sample_Size"], 
                               var_name="Frequency",
                               value_name="Value")
new_readings=readings_converted.drop(['Date_Reading','Start_Freq','Stop_Freq','Freq_Div','Sample_Size'], axis=1)
new_readings
```

As you can see, this code generates a very large dataframe with 1233600 rows and 3 columns. Due to the size of the dataframe, the follow code can run quite slowly depending on the specifications of the PC you are running this on. The first graph gives an overall reading where the Freqency is along the X-Axis and the decible level is on the Y-Axis. What you are looking for is "bumps" in the readout. When the radio telescope hasn't picked up anything, the readout here is quite symmetrical. 

```python
fig=px.line(new_readings,x="Frequency",y="Value",title="Readings")
fig.update_traces(hovertemplate = None,
                  hoverinfo = "skip")
fig.show()
```

This graph looks at the specific freqencies of the hydrogren line. A reading from the hydrogen line is between 1.420400Ghz and 1.420406Ghz. In this instance, you are looking for something that resembles a "dome type" structure. 

```python
new_readings['Frequency']=pd.to_numeric(new_readings['Frequency'])
hydrogen_line=new_readings.query("Frequency > = 1420398437.5 and Frequency < = 1420406250")
fig=px.line(hydrogen_line,x="Time_Reading",y="Value",title="Readings", color="Frequency")
fig.show()
```
